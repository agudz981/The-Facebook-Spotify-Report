```python
keys = pd.read_csv('spotify_key', header = 0, sep = ",")
```

```python
keys
```

```python
appid = keys['Client ID'][0]
```

```python
appsecret = keys['Client_Secret'][0]
```

<span style ="font-family:Arial">To get the spotify data, I need access to the data. Through keys, its possible to gain access to the archive. The keys also helps determine what exactly the data from the keys is through Client ID and Client_Secret and as variables to use later in the report.</span>

```python
client_cred = base64.b64encode(str(appid + ":" + appsecret).encode('ascii'))
```

```python
client_cred
```

```python
header = {'Authorization': 'Basic {}'.format(client_cred.decode('ascii'))}
```

<span style ="font-family:Arial">Next a client_cred variable is created. By using b64.encode, the previous variables, appid and appsecrect, which contains the two categories from the Spotify key are connected.</span> 

```python
payload = {'grant_type': 'client_credentials'}
access_token_url = 'https://accounts.spotify.com/api/token'
```

```python
response = requests.post(access_token_url, headers = header, data = payload)
```

<span style ="font-family:Arial">By creating a variable linked with an access token I can now request certain aspects from the Spotify data.</span> 

```python
response.status_code
```

<span style ="font-family:Arial">To make sure my request was functioning I coded in a status_code to see if its working. The 200 means the request was successfully taken by the API.</span>

```python
response.text
```

```python
session_token = response.json()['access_token']
```

```python
session_header = {'Authorization' : 'Bearer {}'.format(response.json()['access_token'])}
```

```python
album_feature_endpoint = "https://api.spotify.com/v1/albums/{}".format('5CTygC3aONv7l0klY4k3hc')
```

<span style ="font-family:Arial">Now with the final token created and an endpoint to get to the data, albums can now be searched from the archive.</span>

```python
anthem_id = '2Csf9jkWwwusFvIW7DppX7'
```

```python
anthem_id = '2Csf9jkWwwusFvIW7DppX7'
album_tracks_endpoint = "https://api.spotify.com/v1/albums/{}/tracks".format(anthem_id)
anthem_tracks = requests.get(album_tracks_endpoint, headers = session_header)
print(anthem_tracks.status_code)
anthem_dict = json.loads(anthem_tracks.text)
anthem_track_ids = [x['id'] for x in anthem_dict['items']]
anthem_track_strings = ','.join(str(x) for x in anthem_track_ids)
print(anthem_track_strings)
track_features_endpoint = "https://api.spotify.com/v1/audio-features?ids={}".format(anthem_track_strings)
```

<span style ="font-family:Arial">The first set of graphs I've created are histograms, pitting the two albums against each other to a certain audio feature (danaceability and energy). From the danceability graph, a majority of Anthem tracks ranked around 0.35 while the majority of tracks from Hang Ups were ranked 0.55.
```python
fig3, ax3 = plt.subplots()
ant_df = af[af['album'] == 'anthem']
hang_df = af[af['album'] == 'hang_ups']
ax3.scatter(ant_df['danceability'], ant_df['energy'], label ='Anthem', color = 'orange')
ax3.scatter(hang_df['danceability'], ant_df['energy'], label ='Hang Ups', color = 'black')
ax3.set_xlabel('Danceability')
ax3.set_ylabel('Energy')
ax3.set_title('Anthem & Hang Ups: Danceability & Energy Compared')
ax3.legend()
plt.show()
```

<span style ="font-family:Arial">The next set of graphs created were scatter plots to see the relationships between danceability and energy between each album. What's interesting is that on both plots, the track that is the most danceable has lower energy. While tracks with higher energy are less danceable. This could be a possible correlation; the less energy a song the more danceable it is and the more energy a song has the less dancable it is.</span>

```python
fig4, ax4 = plt.subplots()
ant_val = af[af['album'] == 'anthem']['danceability']
hang_val = af[af['album'] == 'hang_ups']['danceability']
ant_x = [1] * len(ant_val)
hang_x = [2] * len(hang_val)
ant_x_jitter = np.random.uniform(-0.10, .10, len(ant_x)) 
hang_x_jitter = np.random.uniform(-0.10, .10, len(hang_x)) 
ax4.scatter(ant_x + ant_x_jitter, ant_val, alpha = 0.2, s= 50, color = 'orange')
ax4.scatter(hang_x + hang_x_jitter,hang_val, alpha = 0.2, s= 50, color = 'black')
ax4.set_xlim(0.5, 2.5)
ax4.boxplot([ant_val, hang_val])
ax4.set_xticks([1,2], ['Anthem', 'Hang Ups'])
ax4.set_ylabel('Danceability')
ax4.set_title('Anthem & Hang Ups: Danceability Compared')
plt.show()
```

<span style ="font-family:Arial">The next graph is a combo between a scatter plot and a box plot to show the median of tracks danceability. While the histograms from above were able to convey the median, the boxplot is able to visually convey the median more precisely.
